```python
import os, json

stride = [0, 5]
model  = ['dbpn','rbpn']
cases  = ['cvd_wse2_hp_gmft','s76ft','s76bt']
for m in model:
    for c in cases:
        os.makedirs(f'./params_{m}/{c}', exist_ok=True)
    for _s in stride:
        # base
        param = {
            'tag':f'cvd_wse2_hp/{m}/base/s0{_s}_2x',
            'pair_path':'./dataset/cvd_wse2_hp_pt_2x.json',
            'logging_interval':50,
            'epochs':150,
            'from_scratch': False,
            'channel_stride':_s,
        }
        if m == 'dbpn':
            param.update({
                'pretrained_model':'./dbpn/models/DBPN-RES-MR64-3_2x.pth',
                'batch_size':36
            })
        elif m == 'rbpn':
            param.update({
                'pretrained_model':'./rbpn/models/RBPN_2x.pth',
                'batch_size':16
            })

#        print(json.dumps(param, indent=4))
        with open(f'./params_{m}/{cases[0]}/s0{_s}_2x.json','w') as f:
            json.dump(param, f, indent=4)

        param.update({
            'tag':f's76_cvd_wse2/{m}/finetune/s0{_s}_2x',
            'pair_path':'./dataset/cvd_wse2_hp_s76_2x.json',
            'logging_interval':10,
            'epochs':100,
            'data_augmentation': False,
        })
        if m == 'dbpn':
            param.update({'batch_size':64})
        elif m == 'rbpn':
            param.update({'batch_size':32}) 

#        print(json.dumps(param, indent=4))
        with open(f'./params_{m}/{cases[1]}/s0{_s}_2x.json','w') as f:
            json.dump(param, f, indent=4)

        param.update({
            'tag':f's76_cvd_wse2/{m}/basetune/s0{_s}_2x',
            'pretrained_model':f'/mnt/MODELS/super_res/pl/cvd_wse2_hp/{m}/base/s0{_s}_2x/00150.model.th'
        })

#        print(json.dumps(param, indent=4))
        with open(f'./params_{m}/{cases[2]}/s0{_s}_2x.json','w') as f:
            json.dump(param, f, indent=4)

```
